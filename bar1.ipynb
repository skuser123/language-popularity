```python
import matplotlib.pyplot as plt
```

```python
l=['c++','python','java','c','javascript']
year=['2000','2003','2004','2005','2006']
u=[40,50,70,60,40]
c=['g','r','b','y','black']
plt.bar(year,u,width=0.2,color=c,label=l)
plt.legend(loc='upper right',fontsize=6)
plt.xlabel('Years')
plt.ylabel('Popularity in (%)')
plt.grid(color='b')
plt.show()
```
